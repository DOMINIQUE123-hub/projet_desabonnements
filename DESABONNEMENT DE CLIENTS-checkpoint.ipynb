```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
```

```python
# chargement du datasabset
valeurs_manquantes = ['Unknown']
data = pd.read_csv('DEPART_CLIENT.csv', na_values = valeurs_manquantes, sep = ';')
data.head()
```

```python
# Renommer les colonnes
data = data.rename(columns =  {'CLIENTNUM': 'ID_CLIENT',
       'Attrition_Flag':'ACTIVITE_CLIENT',
       'Customer_Age':'AGE',
       'Gender':'SEXE',
       'Dependent_count':'NBRE_PERS_A_CHARGE',
       'Education_Level':'NIVEAU_ETUDE',
       'Marital_Status':'STATUT_MATRI',
       'Income_Category':'CATEGO_REVENU_ANNUEL',
        'Card_Category':'TYPE_DE_CARTE',
       'Months_on_book':'PERIOD_RELATIO_BANQUE',
       'Total_Relationship_Count':'NBRE_TOTAL_PRODUITS_CLIENT',
       'Months_Inactive_12_mon':'NBRE_MOIS_INACTIV',
       'Contacts_Count_12_mon':'NBRE_MOIS_INACTIV_12',
       'Credit_Limit':'LIMITE_CREDIT',
       'Total_Revolving_Bal':'SOLDE_RENOUVELABLE',
       'Avg_Open_To_Buy':'LIGNE_CREDIT_ACHAT',
       'Total_Amt_Chng_Q4_Q1':'CHANGEMENT_MONTANT',
       'Total_Trans_Amt':'TOTAL_TRANSAC',
       'Total_Trans_Ct':'NBRE_TOTAL_TRANSAC',
       'Total_Ct_Chng_Q4_Q1':'CHANG_NBRE_TRANSAC',
       'Avg_Utilization_Ratio':'TAUX_UTILI_MOYEN'})
```

```python
data.head()
```

```python
# Recodage des variables ordinales
data['CATEGO_REVENU_ANNUEL'].replace(['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +'], [0, 1, 2, 3, 4], inplace = True)
data['NIVEAU_ETUDE'].replace(['Graduate', 'High School', 'Uneducated', 'College', 'Post-Graduate', 'Doctorate'], [0, 1, 2, 3, 4, 5], inplace = True)
```

```python
data.head()
```

```python
# Recodage des autres variables qualitatives
data_nomi = data[['SEXE','STATUT_MATRI', 'TYPE_DE_CARTE']]
data_nomi = pd.get_dummies(data_nomi, drop_first = True)
```

```python
data.drop(['SEXE','STATUT_MATRI', 'TYPE_DE_CARTE'], axis = 1, inplace = True)
```

```python
data
```

```python
# Encodage de la variable cible
le = LabelEncoder()
data['ACTIVITE_CLIENT'] = le.fit_transform(data['ACTIVITE_CLIENT'])
```

```python
 data.drop(['ID_CLIENT'], axis = 1, inplace = True)
```

```python
# Normalisation des variables quantitatives
data_quanti = data[['AGE', 'PERIOD_RELATIO_BANQUE', 'LIMITE_CREDIT', 'SOLDE_RENOUVELABLE', 'LIGNE_CREDIT_ACHAT', 'CHANGEMENT_MONTANT', 'TOTAL_TRANSAC', 'NBRE_TOTAL_TRANSAC', 'CHANG_NBRE_TRANSAC', 'TAUX_UTILI_MOYEN']]
stand = StandardScaler()
data_num_stand = pd.DataFrame(stand.fit_transform(data_quanti), columns = data_quanti.columns)
```

```python
data.drop(data_num_stand, axis = 1, inplace  = True)
```

```python
data
```

```python
# Vérification de l'existence de valeurs manquantes
data.isna().sum().any()
```

```python
# Concatenation pour l'obtention du dataset final
df = pd.concat([data_nomi,data_num_stand,data[['ACTIVITE_CLIENT', 'NBRE_PERS_A_CHARGE', 'NIVEAU_ETUDE', 'CATEGO_REVENU_ANNUEL', 'NBRE_TOTAL_PRODUITS_CLIENT', 'NBRE_MOIS_INACTIV', 'NBRE_MOIS_INACTIV_12']]], axis = 1)
```

```python
df
```

```python
# Pourcentage de valeurs manquantes par variables
((df.isna().sum()/df.shape[0])*100).sort_values(ascending = False)
```

```python
# Suppression des valeurs manquantes vu que leur pourcentage est faible
df.dropna(axis = 0,inplace = True)
```

```python
df.shape
```

```python
y = df['ACTIVITE_CLIENT']    # variable cible
X = df.drop(['ACTIVITE_CLIENT'], axis = 1)
```

```python
# partage du dataset pour entrainement
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(X_test.shape)
```

## MODELISATION AVEC LE MODELE LOGISTIQUE

```python
# entrainement du modèle de régression logistique
modele_logi = LogisticRegression()
modele_logi.fit(X_train, y_train)
modele_logi.score(X_train, y_train)
```

```python
modele_logi.score(X_test, y_test)
```

## MODELISATION AVEC LE MODELE DE BAGGING

```python
# entrainement du modèle avec le bagging
modele_bag = BaggingClassifier(base_estimator = LogisticRegression(), n_estimators=100)
modele_bag.fit(X_train, y_train)
```

```python
modele_bag.score(X_train, y_train)
```

```python
modele_bag.score(X_test, y_test)
```
